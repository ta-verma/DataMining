## Violations

```
#Violation 1
def species_check(f):
  x = []
  for i in f.index:
    if(f["Species"][i] in ["setosa","versicolor","virginica"]):
      x.append(True)
    else:
      x.append(False)
  x = np.array(x)
  if (False in x):
    print("Violation : Invalid species name")
    print(str(len(x) - np.sum(x)) + " violations")
  else:
    print("No Violation")
  return x
```

```
species_check(df)
```

```
#Violation 2
def all_positive(f):
  x = []
  for i in f.index:
    if(f["Sepal.Length"][i]>0 and f["Sepal.Width"][i]>0 and f["Petal.Length"][i]>0 and f["Petal.Width"][i]>0):
      x.append(True)
    else:
      x.append(False)
  x = np.array(x)
  if (False in x) :
    print("Violation : Non-positive values present")
    print(str(len(x) - np.sum(x)) + " violations")
  else :
    print("No Violation")
  return x
```

```
all_positive(df2)
```

```
#Violation 3
def check_petal(f):
  a = np.array(f["Petal.Length"]>(2*f["Petal.Width"]))
  if (False in a) :
    print("Violation : Petal Length is less than twice of Petal Width in some places")
    print(str(len(a) - np.sum(a)) + " violations")
  else :
    print("No Violation")
  return a
```

```
check_petal(df)
```

```
#Violation 4
def sepal_check(f):
  a = np.array(f["Sepal.Length"]<=30)
  if (False in a) :
    print("Violation : Sepal Length is greater than 30 cm in some places")
    print(str(len(a) - np.sum(a)) + " violations")
  else :
    print("No Violation")
  return a
```

```
species_check(df)
```

```
#Violation 5
def sepal_petal_check(f) :
  a = np.array(f["Sepal.Length"]>f["Petal.Length"])
  if (False in a) :
    print("Violation : Sepal length is greater than petal length in some places")
    print(str(len(a) - np.sum(a)) + " violations")
  else :
    print("No Violation")
  return a
```

```
sepal_petal_check(df)
```

##Summarized Violation

```
print("Summarized result :\n")
print("Violation : Species")
v1 = species_check(df)
print()
v2 = sepal_check(df)
print()
v3 = all_positive(df2)
print()
v4 = check_petal(df)
print()
v5 = sepal_check(df)
print()
v6 = sepal_petal_check(df)
```
