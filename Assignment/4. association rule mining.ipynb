```
import pandas as pd
import numpy as np
from mlxtend.frequent_patterns import apriori, association_rules 
```

```
%matplotlib inline
```

#Groceries DataSet

```
df = pd.read_csv('https://raw.githubusercontent.com/ta-verma/DataSets/master/groceries.csv', sep='\n', header=None)
df.head()
```

##Pre-processing DataSet

```
df = df[0].str.split(',', expand=True)
df.fillna('')
df.head()
```

```
df1 = df.stack().str.get_dummies().sum(level=0)
df1.head()
```

###Frequent items

```
frequent_items = apriori(df1, min_support = 0.05, use_colnames=True)
frequent_items
```

###Association Rule

```
rules = association_rules(frequent_items, min_threshold=0.075)
rules = rules.sort_values(['confidence'], ascending= False)
rules
```

###Apriori

```
frequent_items = apriori(df1, min_support = 0.06, use_colnames=True)
frequent_items
```

```
rules = association_rules(frequent_items, min_threshold=0.06)
rules = rules.sort_values(['confidence'], ascending= False)
rules
```
